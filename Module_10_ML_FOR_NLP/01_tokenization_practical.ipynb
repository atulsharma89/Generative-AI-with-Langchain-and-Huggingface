```python
import nltk
```

```python
nltk.download('all')
```

```python
corpus="""Welocme to world. 
Lets learn AI to become expert"""
```

```python
#Tokenization

#paragraph --> sentences

from nltk.tokenize import sent_tokenize
```

```python
documents = sent_tokenize(corpus)
```

```python
type(documents)
```

```python
for sent in documents:
    print(sent)
```

```python
#Tokenization

#Convert a paragraph into words
#convert sentences to words

from nltk.tokenize import word_tokenize
```

```python
#word tokenization on paragrah
word_tokenize(corpus)
```

```python
#word tokenization on sentences

for sent in documents:
    print(word_tokenize(sent))
```

```python
#for punctuations in corpus

corpus="""Welocme to world. 
Let's learn AI to become expert"""
```

```python
from nltk.tokenize import wordpunct_tokenize
```

```python
wordpunct_tokenize(corpus)
```

```python
#TreeBank word toKENIZER: fUL STOP WILL not be treated as separate 

from nltk.tokenize import TreebankWordTokenizer

token=TreebankWordTokenizer()

tk = token.tokenize(corpus)

tk
```
